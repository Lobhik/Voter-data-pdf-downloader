```python
#importing webdriver from selenium
from selenium import webdriver
import pytesseract
from PIL import Image, ImageEnhance
from selenium.webdriver.common.by import By
import urllib.request
import pandas as pd
import time
```

```python
pytesseract.pytesseract.tesseract_cmd = r'C:\\Program Files\\Tesseract-OCR\\tesseract.exe'
```

```python
df = pd.read_excel('Voter_pdf_list.xlsx')
df.head()
```

```python
link='link'
msg = True
print(link)
for index in range(len(df)):
    link = df.iloc[index , 0]
    # open browser
    download_dir = "C:\\Lobhik"  # for linux/*nix, download_dir="/usr/Public"
    options = webdriver.ChromeOptions()

    profile = {"plugins.plugins_list": [{"enabled": False, "name": "Chrome PDF Viewer"}], # Disable Chrome's PDF Viewer
               "download.default_directory": download_dir , "download.extensions_to_open": "applications/pdf"}
    options.add_experimental_option("prefs", { "plugins.always_open_pdf_externally": True})
    browser = webdriver.Chrome('C:\\Users\\computer\\Program\\chromedriver_win32\\chromedriver', chrome_options=options)  # Optional argument, if not specified will search path.
    print(link)
        #browser = webdriver.Chrome("C:\\Users\\computer\\Program\\chromedriver_win32\\chromedriver")
    #opening link in the browser
    while msg:
        
        browser.get(link)

        # get the image source
        img = browser.find_element(By.ID, 'Image2')

        # taking the image screenshot
        scr = img.screenshot('screenshot.png')
        scr2 = Image.open('screenshot.png')
        enhancer = ImageEnhance.Brightness(scr2)
        factor = 1.5 #brightens the image
        im_output = enhancer.enhance(factor)
        im_output.save('brightened-image.png')
        #converting image to text
        text = pytesseract.image_to_string(im_output)
        print(text)
        #inputing  captcha varification code
        verification = browser.find_element(By.ID, 'txtVerificationCode')
        verification.send_keys(text)
        error_msg = browser.find_element(By.ID, 'lblCaptchaMessage')
        print(error_msg.is_displayed())
        var = error_msg.is_displayed()
        #finding and clicking submit button using id 
        btn = browser.find_element(By.ID, 'btnSubmit')
        print(link)
        btn.click()
        #wait for 10 sec
        time.sleep(10)

        if var == True:
            print("\nerror continue")
            continue

        else:
            print("\nsuccessful")
            browser.close()
            break
            
        #closing browser
            
```
